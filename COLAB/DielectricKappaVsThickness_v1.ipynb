```python
import numpy as np
import matplotlib.pyplot as plt

# Define the thickness range from 0.1 Å to 100 Å for the original graph
thickness_A_original = np.linspace(0.1, 100, 1000)  # in Angstroms
thickness_nm_original = thickness_A_original / 10  # convert Å to nm for the calculations

# Parameters given in the original figure
eps_b = 95  # Static dielectric constant of the bulk material
eps_inf = 5  # High-frequency dielectric constant
K_f = 8.18  # Given parameter, probably related to the Fermi wave vector (in Å^-1)
t_ox = thickness_A_original  # Thickness is represented by 'a' in the original equation

# Equation from the original figure
# Epsilon as a function of the radius (which we are treating as the thickness)
epsilon_original = 1 + ((eps_b - 1) / (1 + (eps_inf * 2 * np.pi / (K_f * t_ox))**2))

# Original figure has two y-axes, we will just plot the dielectric constant for now

plt.figure(figsize=(10, 6))
plt.plot(thickness_A_original, epsilon_original, label='Dielectric Constant ($\kappa$)')

equation_text = r"$\kappa_{ox} = 1 + \frac{\kappa_b - 1}{1 + \left(\frac{\kappa_\infty 2\pi}{K_f t_{ox}}\right)^2}$"
# + "\n" + r"where $\kappa_b = 95$, $\kappa_\infty = 5$, $K_f = 8.18$ Å$^{-1}$"
plt.text(40, 40, equation_text, fontsize=16, bbox=dict(facecolor='white', alpha=0.8))

plt.xlabel('Oxide Thickness [Å]')
plt.ylabel(r'Dielectric Constant ($\kappa$)')
plt.title('Dielectric Constant vs. Thickness')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()


# Redraw the plot with the modified formula text, replacing 'a' with 't_ox' (oxide thickness)

plt.figure(figsize=(10, 6))
plt.semilogx(thickness_A_original, epsilon_original, label='Dielectric Constant (ε)')

# Modify the equation to use 't_ox' instead of 'a'
equation_text = r"$\kappa_{ox} = 1 + \frac{\kappa_b - 1}{1 + \left(\frac{\kappa_\infty 2\pi}{K_f t_{ox}}\right)^2}$"
plt.text(0.10, 40, equation_text, fontsize=18, bbox=dict(facecolor='white', alpha=0.8))

plt.xlabel('Oxide Thickness [Å]')
plt.ylabel('Dielectric Constant ($\kappa$))')
plt.title('Dielectric Constant vs. Thickness (Log-Scale)')
plt.legend(loc='lower right')
plt.grid(True, which="both", ls="--")
plt.show()

```
